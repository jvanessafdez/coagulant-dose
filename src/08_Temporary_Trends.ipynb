# ANÁLISIS DE TENDENCIAS TEMPORALES

```python
# Importación de librerías necesarias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_edit/df_edit_with_met.csv',sep=',', encoding='latin-1', index_col=0, low_memory=False, parse_dates=['FECHA'])
```

```python
df['Año'] = df['FECHA'].dt.year
```

## Años 2013-2015

```python
def trend(start_year,end_year,parameter):
    plt.figure(figsize=(12, 6))
    for year in range(start_year, end_year):
        data_year = df[df['Año'] == year]
        plt.plot(data_year['FECHA'], data_year[parameter], label=year)

    plt.xlabel('FECHA')
    plt.ylabel(parameter)
    plt.title(f'TENDENCIA TEMPORAL DE {parameter}')
    plt.legend(title='Año')
    plt.show()
```

### Agua Cruda

```python
trend(2013,2016,'AGUA_CRUDA_P.H')
```

```python
trend(2013,2016,'AGUA_CRUDA_NTU')
```

```python
trend(2013,2016,'AGUA_CRUDA_COLOR')
```

```python
trend(2013,2016,'CAUDAL')
```

En el análisis de los parámetros de agua cruda, se observa una tendencia definida en los indicadores de turbiedad y color desde mayo hasta septiembre. Durante estos meses, se registra una disminución notable en los valores de estos parámetros. Sin embargo, se evidencia un incremento en los mismos indicadores durante los meses de noviembre y febrero. Esta fluctuación cíclica muestra la variabilidad estacional inherente a estos parámetros de calidad del agua.

### Agua Tratada

```python
trend(2013,2016,'AGUA_TRATADA_P.H')
```

```python
trend(2013,2016,'AGUA_TRATADA_NTU')
```

```python
trend(2013,2016,'AGUA_TRATADA_COLOR')
```

```python
trend(2013,2016,'AGUA_TRATADA_CLORO')
```

En el caso del agua tratada, la tendencia previamente observada no se manifiesta de manera evidente. Sin embargo, en el parámetro de turbiedad, se logra percibir cierta indicación de esta tendencia, aunque no es tan clara como en el agua cruda.

### Datos Meteorológicos

```python
trend(2013,2016,'VEL_VIENTO')
```

Existe una correlación inversa notable entre la velocidad del viento y los parámetros de turbiedad y color. Específicamente, durante los meses en los que estos parámetros disminuyen, la velocidad del viento tiende a incrementar, y viceversa. Esta dinámica sugiere una interacción climática significativa que afecta a la calidad del agua.
