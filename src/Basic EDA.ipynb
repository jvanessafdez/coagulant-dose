# Análisis exploratorio de datos básico

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('./data/dataframe.csv',sep=',', encoding='latin-1', header=[0, 1], index_col=0)
dataframe
# prueba
```

```python
dataframe.info()
```

```python
HEADERS = dataframe.columns
HEADERS
```

```python
# Crear un mapa de nulos
plt.figure(figsize=(12, 8))
sns.heatmap(dataframe.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.xticks(ticks=np.arange(0.5, len(HEADERS)), labels=HEADERS, rotation=90)
plt.title('Missing Values Map')
plt.show()
```

```python
dataframe[('FECHA', 'Unnamed: 1_level_1')] = pd.to_datetime(dataframe[('FECHA', 'Unnamed: 1_level_1')])
```

```python
dataframe['YEAR'] = dataframe[('FECHA', 'Unnamed: 1_level_1')].dt.year
year_clasification = dataframe.groupby('YEAR').agg(lambda x: x.notnull().sum())
year_clasification
```
